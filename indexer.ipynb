```python
def find_word_timestamps(json_output, target_word):
    transcript = json.loads(json_output)
    words = transcript["words"]
    timestamps = []

    for word_data in words:
        word = word_data["word"]
        if word.lower() == target_word.lower():
            start = word_data["start"]
            # end = word_data["end"]
            timestamps.append((start))

    return timestamps
```

```python
target_word = "youtube"
timestamps = find_word_timestamps(formatted_output1, target_word)

print(f"Timestamps for the word '{target_word}':")
for start in timestamps:
    print(f"Start: {start}")
```

```python
def find_word_timestamps_segment(json_output, target_word):
    data = json.loads(json_output)
    timestamps = []

    for segment in data["segments"]:
        text = segment["text"]
        if target_word.lower() in text.lower():
            start_time = segment["start"]
            end_time = segment["end"]
            timestamps.append((start_time, end_time))

    return timestamps
```

```python
target_word1 = 'Youtube'
timestamps1 = find_word_timestamps_segment(formatted_output2, target_word1)

print(f"Timestamps for the word '{target_word1}':")
if timestamps1:
    for start, end in timestamps1:
        print(f"Start: {start}, End: {end}")
else:
    print("No occurrences found.")
```
